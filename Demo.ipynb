```julia
#############################################################
# Simple Plotting Example in Julia using the PyPlot Package #
# Written by: Emad Sadeghi                                  #
#############################################################

using PyPlot 

x = linspace(-10,10,100) # returns an array of 100 evenly spaced numbers over the interval [-10, 10].

y1 = x.^2
y2 = 0*x + 8 # line y = 8
y3 = 0*x + 4 # line y = 4

plot(x, y1, "b", linewidth = 3)
hold 
plot(x, y2, "r--", linewidth = 2)
hold 
plot(x, y3, "r--", linewidth = 2)

####################
# Alternatively, you could have written the following: 
# plot(x, y1, "b", linewidth = 3, x, y2, "r--", linewidth = 2, x, y3, "r--", linewidth = 2)
####################

xlim([-5, 5]) # limits the x values to range between -5 and 5
ylim([-1, 10]) # limits the y values to range between -1 and 10 

####################
# Alternatively, you could have written the following: 
# axis([-5,5,-1,10])
####################

grid() # adds grid lines 
xlabel("x axis") 
ylabel("y axis")
title("Demo Plot")

fill([-2, 2,2, -2], [4,4,8, 8], color= "lightgreen")
# In general, if you want to fill in a polygon specified by (x1, y1), (x2, y2), ..., (xn, yn), do the following: 
# fill([x1, x2, ..., xn], [y1, y1, ..., yn], color = "name_your_color_here")
```
